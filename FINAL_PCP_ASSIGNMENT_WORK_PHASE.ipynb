```python
#  A function for computing the average age, modal age, median age of those whose heart failure resulted in death.

def compute_avg_modal_median_age_death(data):
# 'DEATH_EVENT' is at index 12
    death_data = [row for row in data if int(row[12]) == 1]  
    
    # To get the ages of those who died for'age' at index 0
    ages = [float(row[0]) for row in death_data]  
    
    # To compute age statistic
    avg_age = sum(ages) / len(ages) if ages else None
    modal_age = max(set(ages), key=ages.count) if ages else None
    sorted_ages = sorted(ages)
    n = len(sorted_ages)
    median_age = (sorted_ages[n//2] + sorted_ages[(n-1)//2]) / 2 if ages else None
    
    # Save results in a list
    results = [["Average Age of death", avg_age],
               ["Modal Age of death", modal_age],
               ["Median Age of death", median_age]]
    try:
        # Save the result to a csv file
        with open("pcp1_results.csv", mode='a', newline='') as result_file:
            writer = csv.writer(result_file)
            writer.writerows(results)
    except Exception as e:
            print(f"Error saving results to CSV: {e}")
            
    return results
```

```python
#  A function for computing the average time taken for those whose heart failure did not result in death.

def compute_average_time_for_survival(data):
    # For 'DEATH_EVENT' at index 12
    survival_data = [row for row in data if int(row[12]) == 0]  
    
    # To get time for survival for 'time' is at index 11
    times_for_survival = [float(row[11]) for row in survival_data]  
    
    # Compute average time for survival
    average_time_for_survival = sum(times_for_survival) / len(times_for_survival) if times_for_survival else None
    
    # Save results in a list
    results = [["Average Time Taken for survival", average_time_for_survival]]
    
    try:
        # Save the result to a csv file
        with open("pcp2_results.csv", mode='a', newline='') as result_file:
            writer = csv.writer(result_file)
            writer.writerows(results)
    except Exception as e:
            print(f"Error saving results to CSV: {e}")
            
    return average_time_for_survival


```

```python
#  A function to compute median age, average age, and modal age of those with high blood pressure.

def compute_age_statistic_high_blood_pressure(data):
    # For 'high_blood_pressure' at index 5
    high_bp_data = [row for row in data if int(row[5]) == 1]  

    # To get the age values for those with high blood pressure assuming 'age' is at index 0
    ages_high_bp = [float(row[0]) for row in high_bp_data]  
    
    # To compute median age, average age, and modal age
    median_age = sorted(ages_high_bp)[len(ages_high_bp) // 2] if ages_high_bp else None
    average_age = sum(ages_high_bp) / len(ages_high_bp) if ages_high_bp else None
    modal_age = max(set(ages_high_bp), key=ages_high_bp.count) if ages_high_bp else None

    # Store results in a list
    results = [['Median Age of Those with High Blood Pressure', median_age],
               ['Average Age of Those with High Blood Pressure', average_age],
               ['Modal Age of Those with High Blood Pressure', modal_age]]
    try:
        # Save the results to a CSV file
        with open('pcp3_results.csv', mode='a', newline='') as result_file:
            writer = csv.writer(result_file)
            writer.writerows(results)
    except Exception as e:
            print(f"Error saving results to CSV: {e}")
            
    return median_age, average_age, modal_age


```

```python
#  A function for computing the compute median age, average age, and modal age of those with diabetes

def compute_age_statistic_diabetes(data):
    # For 'diabetes' at index 3
    diabetes_data = [row for row in data if int(row[3]) == 1]  
    
    # To get the age for those with diabetes when 'age' is at index 0
    ages_diabetes = [float(row[0]) for row in diabetes_data]  
    
    # Compute average age, median age, and modal age
    average_age = sum(ages_diabetes) / len(ages_diabetes) if ages_diabetes else None
    sorted_ages = sorted(ages_diabetes)
    median_age = (sorted_ages[len(ages_diabetes) // 2] + sorted_ages[(len(ages_diabetes) - 1) // 2]) / 2 if ages_diabetes else None
    modal_age = max(set(ages_diabetes), key=ages_diabetes.count) if ages_diabetes else None

    # Store results in a list
    results = [['Average Age of those with diabetes', average_age],
               ['Median Age of those with diabetes', median_age],
               ['Modal Age of those with diabetes', modal_age]]
    try:
        # Save the results to a CSV file
        with open('pcp4_results.csv', mode='a', newline='') as result_file:
            writer = csv.writer(result_file)
            writer.writerows(results)
    except Exception as e:
            print(f"Error saving results to CSV: {e}")
            
    return average_age, median_age, modal_age

```

```python
#  A function to determine whether diabetes is linked to smoking and high blood pressure.

def diabetes_linked_to_smoking_and_hbp(data):
    # For 'diabetes' at index 3 and 'high_blood_pressure'at index 5
    diabetes = [int(row[3]) for row in data]
    hbp = [int(row[5]) for row in data]

    # For only those that have diabetes
    diabetes_indices = [i for i, value in enumerate(diabetes) if value == 1]
    
    # To Check if any of those that have diabetes smoke
    smokes = any(hbp[i] == 1 for i in diabetes_indices)
    
    # To check if any of those with diabetes have high blood pressure
    hbp = any(hbp[i] == 1 for i in diabetes_indices)
    
    # To check if there is a relationship
    linked = smokes and hbp
   
    results = [
        ["Linkage", linked]
    ]

    try:
        # Save the results to a CSV file
        with open('pcp5_results.csv', mode='a', newline='') as result_file:
            writer = csv.writer(result_file)
            writer.writerows(results)
    except Exception as e:
            print(f"Error saving results to CSV: {e}")
            
    return linked


```

```python
# A function to compute the average serum sodium of those with diabetes

def compute_average_serum_sodium_with_diabetes(data):
    # For only those that have diabetes for 'diabetes' at index 3
    diabetes_data = [row for row in data if int(row[3]) == 1]  

    # To calculate the average serum sodium for 'serum_sodium' at index 8
    serum_sodium = [float(row[8]) for row in diabetes_data]  
    average_serum_sodium = sum(serum_sodium) / len(serum_sodium) if serum_sodium else 0

    results = [['Average Serum Sodium of those with diabetes', average_serum_sodium]]
    try:
        # Save the results to a CSV file
        with open('pcp6_results.csv', mode='a', newline='') as result_file:
            writer = csv.writer(result_file)
            writer.writerows(results)
    except Exception as e:
            print(f"Error saving results to CSV: {e}")
            
    return average_serum_sodium

```

```python
#  A function to determine if anaemia is linked to smoking or not.

def is_anaemia_linked_to_smoking(data):
    # For'anaemia' at index 1 and 'smoking' at index 10
    anaemia = [int(row[1]) for row in data]
    smoking = [int(row[10]) for row in data]

    # To check for correlation 
    n = len(anaemia)
    anaemia_mean = sum(anaemia) / n
    smoking_mean = sum(smoking) / n
    numerator = sum((anaemia[i] - anaemia_mean) * (smoking[i] - smoking_mean) for i in range(n))
    anaemia_std_dev = (sum((anaemia[i] - anaemia_mean) ** 2 for i in range(n)) / n) ** 0.5
    smoking_std_dev = (sum((smoking[i] - smoking_mean) ** 2 for i in range(n)) / n) ** 0.5

    correlation = numerator / (anaemia_std_dev * smoking_std_dev)

    if abs(correlation) >= 0.2:
        result = "Anaemia is linked to smoking"
    else:
        result = "Anaemia is not linked to smoking"

    results = [
        ["Anaemia Linked to Smoking", result]
    ]
    try:
        
        # Save the results to a CSV file
        with open('pcp7_results.csv', mode='a', newline='') as result_file:
            writer = csv.writer(result_file)
            writer.writerows(results)
    except Exception as e:
            print(f"Error saving results to CSV: {e}")
    return result


```

```python
#  A function that computes and returns the IQRs (Interquartile Range) of ejection fraction and serum creatinine.

def compute_iqrs(data):
    # For 'ejection_fraction' at index 4 and 'serum_creatinine' is at index 7
    ejection_fraction = [float(row[4]) for row in data]
    serum_creatinine = [float(row[7]) for row in data]

    ejection_fraction_iqr = compute_iqr(ejection_fraction)
    serum_creatinine_iqr = compute_iqr(serum_creatinine)

    results = [
        ["IQR of Ejection Fraction", ejection_fraction_iqr],
        ["IQR of Serum Creatinine", serum_creatinine_iqr]
    ]
    try:
        
        # Save the results to a CSV file
        with open('pcp9_results.csv', mode='a', newline='') as result_file:
            writer = csv.writer(result_file)
            writer.writerows(results)
    except Exception as e:
            print(f"Error saving results to CSV: {e}")
            
    return ejection_fraction_iqr, serum_creatinine_iqr

def compute_iqr(data):
    sorted_data = sorted(data)
    q1 = sorted_data[int(len(sorted_data) * 0.25)]
    q3 = sorted_data[int(len(sorted_data) * 0.75)]
    iqr = q3 - q1
    
    return iqr


```

```python
#  A function to compute and return the sample variance of creatinine phosphokinase and serum sodium.

def compute_sample_variances(data):
    # For'creatinine_phosphokinase (cp)' at index 2 and 'serum_sodium (ss)' is at index 8
    cp = [float(row[2]) for row in data]
    ss = [float(row[8]) for row in data]
    
    cp_variance = compute_sample_variance(cp)
    ss_variance = compute_sample_variance(ss)

    results = [
        ["Sample Variance of Creatinine Phosphokinase", cp_variance],
        ["Sample Variance of Serum Sodium", ss_variance]
    ]

    try:
        # Save the results to a CSV file
        with open('pcp10_results.csv', mode='a', newline='') as result_file:
            writer = csv.writer(result_file)
            writer.writerows(results)
    except Exception as e:
            print(f"Error saving results to CSV: {e}")
            
    return cp_variance, ss_variance

def compute_sample_variance(data):
    n = len(data)
    mean = sum(data) / n
    variance = sum((x - mean) ** 2 for x in data) / n
    
    return variance


```
